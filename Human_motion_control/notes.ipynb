## An example of tasks: take a coffee

- Where is my body? Where is the bar?
> Modalities (vision, proprioception, multimodal integration...), and references frames (target position, arm position)
- How to reach the bar? Where is the cup? Where is my arm?
> Optic ataxia (visuomotor coordination), deafferentation (loss of proprioception)
- How to reach the cup?
> - Choice of a task-space trajectory (path, time course along the path), 
> - Choice of a body-space trajectory (DOF, kinematic redundancy: number of DOF > task-space dimension)
> - Kinematics: coordinate transformation, inverse kinematics is an ill-posed problem

- How to calculate the motor command?
> - Joint torques (desired body-space trajectory), 
> - Dynamics (Direct/inverse transformation (Newton’s law)), 
> - Force distribution (dynamic redundancy), 

- How to interact with the environment?
> - Stable (e.g. firmly grasped, rigid objects)
> - Unstable: tool manipulation, posture and gait control (e.g. using a drill, stick balancing...)

- Is the command correct?
> Identification of errors' origin, and solution finding: feedback

- How to do better at the next trial?
> - Adaptation, motor learning (biomechanical interface: tool, telemanipulation visuomotor transformation, dynamic transformation)
> - Nature of adaptation and learning (temporary vs permanent interferences, learning vs development)
> - Error signals (all or none (success / failure) — quantitative)

## 1. The organization of action main vocabulary

### The degrees of freedom problem

> Redundancy, nonlinearity problems (in task space, body space, muscle space, neural space): 
> - accurate total force production: F-solutions
> - zero total moment of force: M-solutions

> Slowness / Time delays: in afferent sensory information (receptor dynamics, conduction delays)
> In efferent motor commands, *« We live in the past »*

> Noise and its consequences: signal-dependent motor noise

> Nonstationary factors: growth development, ageing, training, muscle fatigue, muscle tears

> Interaction (Rigid objects: load force, grip force...)

> Perturbations / Correction: Motor control is highly flexible in space and time. Error corrections
are needed only if perturbations affect the behavioral goal, they are ignored otherwise. Note that corrective responses do not necessarily return to a desired trajector.

### Adaptation Learning and motor control

- Adaptation: regain former capabilities in altered circumstances (e.g. prism, force fields, ...)
- Learning: change, resulting from practice or a novel experience, in the capability for responding (e.g. piano, golf)

## 2. Computational motor control (main concepts)

#### Motor Invariants 

- Trajectories are point-to-point movements: straight, bell-shaped velocity profiles.

- Scaling laws are the duration and velocity scale with amplitude and load.

- Motor equivalence: Actions are encoded in the central nervous system in terms that are more abstract than commands to specific muscles

- EMG: triphasic pattern during fast movements
